```python
from pathlib import Path
import pandas as pd
import random
import string
from datetime import datetime, timedelta

random.seed(42)
Path('output').mkdir(parents=True, exist_ok=True)

male_last_names = ['Иванов', 'Петров', 'Сидоров', 'Смирнов', 'Кузнецов', 'Попов', 'Васильев', 'Новиков', 'Фёдоров', 'Морозов', 'Волков', 'Алексеев', 'Лебедев', 'Семёнов', 'Егоров', 'Павлов', 'Козлов', 'Степанов', 'Николаев', 'Орлов']
male_first_names = ['Иван', 'Пётр', 'Сергей', 'Алексей', 'Дмитрий', 'Андрей', 'Николай', 'Владимир', 'Михаил', 'Артём', 'Евгений', 'Роман', 'Олег', 'Максим', 'Виктор', 'Денис', 'Антон', 'Павел', 'Кирилл', 'Илья']
male_patronymics = ['Иванович', 'Петрович', 'Сергеевич', 'Алексеевич', 'Дмитриевич', 'Андреевич', 'Николаевич', 'Владимирович', 'Михайлович', 'Артёмович', 'Евгеньевич', 'Романович', 'Олегович', 'Максимович', 'Викторович', 'Денисович', 'Антонович', 'Павлович', 'Кириллович', 'Ильич']

female_last_names = ['Иванова', 'Петрова', 'Сидорова', 'Смирнова', 'Кузнецова', 'Попова', 'Васильева', 'Новикова', 'Фёдорова', 'Морозова', 'Волкова', 'Алексеева', 'Лебедева', 'Семёнова', 'Егорова', 'Павлова', 'Козлова', 'Степанова', 'Николаева', 'Орлова']
female_first_names = ['Анна', 'Мария', 'Елена', 'Ольга', 'Наталья', 'Ирина', 'Татьяна', 'Светлана', 'Екатерина', 'Юлия', 'Алина', 'Ксения', 'Дарья', 'Виктория', 'Полина', 'Анастасия', 'Евгения', 'Людмила', 'Вероника', 'София']
female_patronymics = ['Ивановна', 'Петровна', 'Сергеевна', 'Алексеевна', 'Дмитриевна', 'Андреевна', 'Николаевна', 'Владимировна', 'Михайловна', 'Артёмовна', 'Евгеньевна', 'Романовна', 'Олеговна', 'Максимовна', 'Викторовна', 'Денисовна', 'Антоновна', 'Павловна', 'Кирилловна', 'Ильинична']

def make_name():
    if random.random() < 0.5:
        return f"{random.choice(male_last_names)} {random.choice(male_first_names)} {random.choice(male_patronymics)}"
    return f"{random.choice(female_last_names)} {random.choice(female_first_names)} {random.choice(female_patronymics)}"

def gen_inn(valid=True):
    if valid:
        return ''.join(random.choice(string.digits) for _ in range(random.choice([10, 12])))
    return ''.join(random.choice(string.digits) for _ in range(random.choice([5, 7, 9, 11, 13, 15])))

def gen_passport():
    return f"{random.randint(1000, 9999)} {random.randint(100000, 999999)}"

def gen_phone(valid=True):
    if valid:
        return f"+79{random.randint(100000000, 999999999)}"
    return random.choice([
        f"8{random.randint(1000000, 9999999)}",
        f"+7{random.randint(100, 999)}",
        f"+79{random.randint(10000, 99999)}",
        f"{random.randint(1000000000, 9999999999)}"
    ])

def gen_date():
    start = datetime(2024, 1, 1)
    return (start + timedelta(days=random.randint(0, 850))).strftime('%Y-%m-%d')

rows = []
base_rows = []
for i in range(1, 81):
    row = {
        'client_id': i,
        'full_name': make_name(),
        'inn': gen_inn(valid=(i % 10 != 0)),
        'passport': gen_passport(),
        'phone': gen_phone(valid=(i % 8 != 0)),
        'registration_date': gen_date(),
    }
    base_rows.append(row)
    rows.append(row)

for i in range(81, 101):
    src = random.choice(base_rows[:25])
    row = src.copy()
    row['client_id'] = random.choice([src['client_id'], random.randint(1, 80)])
    if random.random() < 0.5:
        row['full_name'] = make_name()
    if random.random() < 0.5:
        row['inn'] = gen_inn(valid=False)
    if random.random() < 0.5:
        row['phone'] = gen_phone(valid=False)
    row['registration_date'] = gen_date()
    rows.append(row)

df = pd.DataFrame(rows)
out = 'output/clients_test_data_100.csv'
df.to_csv(out, index=False)
```

```python
import pandas as pd
import re
import numpy as np


#  Загрузка данных
df = pd.read_csv('output/clients_test_data_100.csv')



def find_duplicates_by_name_passport(df: pd.DataFrame) -> pd.DataFrame:
    """Возвращает строки-дубликаты по full_name + passport"""
    dup = df.duplicated(subset=['full_name', 'passport'], keep=False)
    return df[dup]

def find_duplicates_by_inn(df: pd.DataFrame) -> pd.DataFrame:

    dup = df.duplicated(subset=['inn'], keep=False)
    return df[dup]

def validate_inn(inn: str) -> bool:

    if not isinstance(inn, str) or not inn.isdigit():
        return False
    if len(inn) == 10:
        coeff = [2, 4, 10, 3, 5, 9, 4, 6, 8]
        total = sum(int(inn[i]) * coeff[i] for i in range(9))
        control = total % 11
        if control == 10:
            control = 0
        return control == int(inn[9])
    return len(inn) == 12

def validate_phone(phone: str) -> bool:
    """Проверка формата +7XXXXXXXXXX (ровно 12 символов, после +7 – 10 цифр)"""
    return bool(re.fullmatch(r'\+7\d{10}', str(phone)))


np.random.seed(42)
client_ids = df['client_id'].unique()
clients_with_accounts = np.random.choice(client_ids, size=int(len(client_ids)*0.8), replace=False)
accounts = []
for cid in clients_with_accounts:
    for _ in range(np.random.randint(1, 4)):
        accounts.append({'account_id': len(accounts)+1, 'client_id': cid})
df_accounts = pd.DataFrame(accounts)

def check_referential_integrity(df_clients: pd.DataFrame, df_accounts: pd.DataFrame) -> pd.DataFrame:

    missing = set(df_clients['client_id']) - set(df_accounts['client_id'])
    return df_clients[df_clients['client_id'].isin(missing)]


```

```python
def classify_errors(row, df_full, accounts_ids):
    errors = []


    inn = str(row['inn']).strip()
    if not (len(inn) in [10, 12] and inn.isdigit()):
        errors.append("Некорректный формат ИНН")

    phone = str(row['phone']).strip()
    if not re.fullmatch(r'\+7\d{10}', phone):
        errors.append("Неверный формат телефона")


    is_duplicate = df_full[(df_full['full_name'] == row['full_name']) &
                           (df_full['passport'] == row['passport'])].shape[0] > 1
    if is_duplicate:
        errors.append("Технический дубликат")

    if row['client_id'] not in accounts_ids:
        errors.append("Отсутствуют открытые счета")

    return ", ".join(errors) if errors else "Ошибок не обнаружено"

clients_with_accounts = df_accounts['client_id'].unique()

report_df = df.copy()


report_df['found_violations'] = report_df.apply(
    lambda r: classify_errors(r, df, clients_with_accounts), axis=1
)


error_report = report_df[report_df['found_violations'] != "Ошибок не обнаружено"]



print(f"Общее количество записей: {len(df)}")
print(f"Количество записей с нарушениями: {len(error_report)}")
print("-" * 30)

error_report[['client_id', 'full_name', 'inn', 'phone', 'found_violations']].head(20)
```

```python

def check_inn(inn):
    if not isinstance(inn, str) or not inn.isdigit():
        return False
    if len(inn) == 10:
        coeff = [2, 4, 10, 3, 5, 9, 4, 6, 8]
        total = sum(int(inn[i]) * coeff[i] for i in range(9))
        control = total % 11
        if control == 10:
            control = 0
        return control == int(inn[9])
    return len(inn) == 12

dup_mask = df.duplicated(subset=['full_name', 'passport'], keep=False)
duplicates = df[dup_mask].copy()

unique_records = []
processed = set()

for idx, row in duplicates.iterrows():
    key = (row['full_name'], row['passport'])

    if key in processed:
        continue


    group = duplicates[(duplicates['full_name'] == key[0]) & (duplicates['passport'] == key[1])]

 
    golden = group.iloc[0].copy()


    valid_inns = group[group['inn'].apply(check_inn)]
    if not valid_inns.empty:
        golden['inn'] = valid_inns.iloc[0]['inn']
    else:
        len_ok = group[group['inn'].astype(str).str.len().isin([10, 12])]
        if not len_ok.empty:
            golden['inn'] = len_ok.iloc[0]['inn']
        else:
            golden['inn'] = group.loc[group['inn'].astype(str).str.len().idxmax(), 'inn']


    valid_phones = group[group['phone'].astype(str).str.match(r'\+7\d{10}')]
    if not valid_phones.empty:
        golden['phone'] = valid_phones.iloc[0]['phone']
    else:
        golden['phone'] = group.loc[group['phone'].astype(str).str.len().idxmax(), 'phone']


    golden['registration_date'] = group['registration_date'].max()

    unique_records.append(golden)
    processed.add(key)


non_duplicates = df[~dup_mask].copy()
result = pd.concat([non_duplicates, pd.DataFrame(unique_records)], ignore_index=True)

print("Уникальных клиентов после объединения дубликатов:", len(result))
print("Было удалено:", len(df) - len(result))
```

```python
import json
from datetime import datetime

audit_log = []

def fix_client_data(client_id, field, old_value, new_value, reason):
    """Функция исправления данных с логированием"""
    audit_log.append({'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        'user': 'data_engineer',
        'action': 'fix_data',
        'client_id': client_id,
        'field': field,
        'old_value': old_value,
        'new_value': new_value,
        'reason': reason})

fix_client_data(15, 'inn', '12345', '123456789012', 'некорректная длина ИНН')
fix_client_data(23, 'phone', '8912345', '+79123456789', 'неправильный формат телефона')
fix_client_data(67, 'registration_date', '2024-01-01', '2025-03-15', 'обновление после верификации')

print("Лог аудита")
for entry in audit_log:
    print(f"[{entry['timestamp']}] {entry['user']}")
    print(f"Клиент {entry['client_id']}, поле: {entry['field']}")
    print(f"{entry['old_value']} → {entry['new_value']}")
    print(f"Причина: {entry['reason']}\n")

with open('output/audit_log.json', 'w', encoding='utf-8') as f:
    json.dump(audit_log, f, ensure_ascii=False, indent=2)
```
